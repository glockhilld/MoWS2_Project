```python
def manu_picks_to_(image, points, points_list, points_index, extent=[[0, 1024], [0, 1024]]):
    counts = 0
    def onpick(event):
        ind = event.ind
        global counts
        counts += 1
        points_list.append(points[:, ][ind])
        points_index.append(ind)
        print(f'onpick scatter no.{counts}:', ind, points[:, ][ind])

    fig, ax = plt.subplots()
    ax.plot(points[:, 1], points[:, 0], 'ro', alpha=.3, lw=1, picker=3)
    ax.imshow(image, 'gray')
    ax.axis('off')
    ax.set_xlim(extent[0])
    ax.set_ylim(extent[1])
    fig.canvas.mpl_connect('pick_event', onpick)
    fig.show()
```

```python
lattice_heavy = feature.blob_dog(nimg, threshold=.055, min_sigma=1, max_sigma=30, overlap=.01)
fig2, ax2 = plt.subplots()
ax2.plot(lattice_heavy[:, 1], lattice_heavy[:, 0], 'ro', alpha=.4, lw=1)
ax2.imshow(img, 'gray')
ax2.axis('off')
fig2.show()
```

```python
media_set = list(whole_set)
l_ind = []
lattice_heavy_list = list(lattice_heavy)
ccounts = 0
for i in lattice_heavy_list:
    for index, element in enumerate(media_set):
        diff = np.sum((element - i)**2)
        if diff > 1e-20:
            #print(diff)
            continue
        else:
            l_ind.append(index)
            ccounts += 1
            #print(index)   
```

```python
lattice_light = np.delete(p.copy(), l_ind, axis=0)
```

```python
def cross_rdf_cal(lattice_heavy, lattice_light, cellPara=[1024,1024], pixelSize=[1,1], bins=2048, unit='nm'):
    cellPara = np.array(cellPara, dtype='float64')
    pixelSize = np.array(pixelSize, dtype='float64')
    #realsize = np.array(cellPara, dtype='float64') * np.array(pixelSize, dtype='float64')
    dr = cellPara[0] / 2 / bins
    r = np.ones(bins) * dr * np.arange(1, bins+1)
    cross_g = np.zeros((bins, 2))
    cross_g[:, 0] = r
    from math import dist 
    for coords in lattice_light:
        for coords2 in lattice_heavy:
            distance = dist(coords, coords2)
            index = int(distance / dr)
            if 0 < index < bins:
                cross_g[index,1] += 2.0

    volume = np.zeros(bins)
    pi = np.pi
    for k in range(bins):
        r1 = k * dr
        r2 = r1 + dr
        s1 = 2 * pi * r1 **2
        s2 = 2 * pi * r2 **2
        volume[k] += s2 - s1
    cross_g[:,1] /= volume
    cross_g[:,0] *= pixelSize[0]
    max_peak = np.argmax(cross_g[:,1])
    max_val = np.max(cross_g[:,1])
    print(f'The highest peak is located at {max_peak * pixelSize[0]} nm with intensity of {max_val}')
    return cross_g
```

```python
cross_g = cross_rdf_cal(lattice_heavy, lattice_light, pixelSize=[0.0178823, 0.0178823], bins=2048)
```
